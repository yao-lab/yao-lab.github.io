```python
import pandas as pd
```

```python
bur = pd.read_csv('./data/bureau.csv')
bur
```

```python
b = bur.groupby(['SK_ID_CURR']).sum()
b
```

```python
target = pd.read_csv('./data/application_train.csv')
target = target[['SK_ID_CURR', 'TARGET']]
target
```

```python
res = pd.merge(b, target, on='SK_ID_CURR')
res
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_days_col(data_set, col):
    plt.figure()
    sns.kdeplot(data_set.loc[data_set['TARGET'] == 0, col], label = 'good')
    sns.kdeplot(data_set.loc[data_set['TARGET'] == 1, col], label = 'bad')
    plt.legend()
    plt.show()
```

```python

plot_days_col(res, 'DAYS_CREDIT')
# DAYS_CREDIT can be our features
```

```python
overdue = res[res['AMT_CREDIT_SUM_DEBT'] > 0]
overdue['TARGET'].value_counts()
res[res['SK_ID_CURR'] == 259355]
overdue[overdue['TARGET']==1]
```

```python
bur_with_target = pd.merge(bur, target, on='SK_ID_CURR')
bad_guys = bur_with_target[bur_with_target['TARGET'] == 1]
bad_guys['CREDIT_TYPE'].value_counts() / bur_with_target['CREDIT_TYPE'].value_counts()
```

```python
credit_type = bur_with_target[['SK_ID_CURR', 'TARGET', 'CREDIT_TYPE']]
credit_type = pd.get_dummies(credit_type)
credit_type = credit_type.groupby(['SK_ID_CURR'], as_index=False).max()
credit_type
```

```python
credit_days = bur_with_target[['SK_ID_CURR', 'TARGET', 'DAYS_CREDIT']]
credit_days[credit_days['TARGET'] == 1]
credit_cnt = credit_days.groupby(['SK_ID_CURR'])['TARGET'].count().reset_index(name='CNT')
credit_days = pd.merge(credit_days, credit_cnt, on='SK_ID_CURR')
credit_days['DAYS_DIFF'] = credit_days.sort_values('DAYS_CREDIT').groupby('SK_ID_CURR')['DAYS_CREDIT'].diff()
credit_days[credit_days['TARGET'] == 1]
```

```python
DAYS_CREDIT_MIN = credit_days.groupby('SK_ID_CURR')['DAYS_CREDIT'].max()
DAYS_DIFF_MIN = credit_days.groupby('SK_ID_CURR')['DAYS_DIFF'].min()
bur_days = pd.merge(DAYS_CREDIT_MIN, DAYS_DIFF_MIN, on='SK_ID_CURR')
plot_days_col(credit_days, 'DAYS_CREDIT')
```

```python
# DEBT
credit_debt = bur_with_target[['SK_ID_CURR', 'TARGET', 'CREDIT_TYPE', 'AMT_CREDIT_SUM', 'AMT_CREDIT_SUM_DEBT', 'AMT_CREDIT_SUM_OVERDUE']]
credit_debt['AMT_CREDIT_SUM_DEBT'] = credit_debt['AMT_CREDIT_SUM_DEBT'].clip(0)

credit_debt['CREDIT_DEBT'] = credit_debt['AMT_CREDIT_SUM_DEBT'] - credit_debt['AMT_CREDIT_SUM']
credit_debt[(credit_debt['TARGET']==0) & (credit_debt['AMT_CREDIT_SUM_OVERDUE']>0)]

credit_debt['IS_OVERDUE'] = credit_debt['AMT_CREDIT_SUM_OVERDUE'].apply(lambda x:1 if x>0 else 0)
is_overdue = credit_debt.groupby('SK_ID_CURR').sum()
type(is_overdue)
```
