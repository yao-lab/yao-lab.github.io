```python
import pandas as pd
import numpy as np
import math
```

```python
data=pd.read_csv('../data/GKX_20201231.csv')
```

```python
data.head()
```

# 1. Select data from 1960 to 2019

```python
data = data[(data['DATE'] >= 19600101) & (data['DATE'] <= 20191231)].reset_index(drop = True)
data
```

```python
data['DATE']=data['DATE']//100 
data = data.rename(columns={'DATE': 'yyyymm'})#Let the DATE become the format "yyyymm"
data
```

# 2. Construct full dataset

```python
macro = pd.read_csv('../data/macro.csv')
macro["Index"] = macro["Index"].apply(lambda x: float(x.replace(",", "")))
macro = macro.loc[(macro['yyyymm']>=196001)&(macro['yyyymm']<202001)]
macro
```

## 2.1 Calculate some macropredictors that are not given directly
       Dividend to price ratio(dp):log(D12)-log(Index)
       Earnings-price ratio(ep):log(E12)-log(Index)
       Term spread(tms):lty-tbl
       Default spread(dfy):BAA-AAA

```python
macro['dp'] = macro['D12'].apply(math.log) - macro['Index'].apply(math.log)
macro['ep'] = macro['E12'].apply(math.log) - macro['Index'].apply(math.log)
macro['tms'] = macro['lty']-macro['tbl']
macro['dfy'] = macro['BAA']-macro['AAA']
#Rename b/m by bm and Keep only 8 needed macropredictors
ned = ['yyyymm', 'dp', 'ep', 'bm', 'ntis', 'tbl', 'tms', 'dfy','svar']
macro = macro.rename(columns = {'b/m': 'bm'})[ned]
```

```python
data = pd.merge(data, macro, how='left', on='yyyymm', suffixes=('', '_macro'))
data
```

## 3. Fill the missing value with cross-sectional median or 0

```python
nonpd = ['yyyymm','RET','SHROUT','mve0','prc','permno'] #Nonpredictors
macropd = ['dp','ep_macro','bm_macro','ntis','tbl','tms','dfy','svar'] #Macroeconomic predictors
sic2 = ['sic2'] #Industrial dummies
lst = nonpd + macropd + sic2
stockpd = [p for p in data.columns if p not in lst] #Stock-level predictors
len(stockpd)
```

```python
period_list = np.unique(data['yyyymm'])
len(period_list)
```

```python
for pd in stockpd:
    c_row = 0
    for period in period_list:
        n_period = data[data['yyyymm'] == period].shape[0]
        #Compute the cross-sectional median of this stock predictor
        median = np.nanmedian(data.loc[c_row:(c_row+n_period-1),pd])
        #Fill missing values with median
        data.loc[c_row:(c_row+n_period-1),pd] = data.loc[c_row:(c_row+n_period-1),pd].replace(np.nan,median)
        c_row = c_row + n_period
```

```python
newdt = data.fillna(0)
```

```python
newdt = newdt.drop(columns=['prc','SHROUT','mve0']) #Drop the predictors that are not used
newdt.to_csv('../data/processed_data.csv', index = False)
```
