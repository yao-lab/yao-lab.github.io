```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
# state = torch.load('./checkpoints/rgstoreTrue_start1993_period8_preds5_lr1e-05_epoch200_bnTrue_xinitTrue_1stout64_layers3_drop0.5_lrelu_mp(2, 1)_flt(5, 3)_dlt(2, 1)_strd(1, 1)_200th.cpt')
# state = torch.load('./checkpoints/rgstoreTrue_start1993_period8_preds20_lr1e-05_epoch200_bnTrue_xinitTrue_1stout64_layers3_drop0.5_lrelu_mp(2, 1)_flt(5, 3)_dlt(2, 1)_strd(1, 1)_200th.cpt')
state = torch.load('./checkpoints/rgstoreTrue_start1993_period8_preds60_lr1e-05_epoch200_bnTrue_xinitTrue_1stout64_layers3_drop0.5_lrelu_mp(2, 1)_flt(5, 3)_dlt(2, 1)_strd(1, 1)_200th.cpt')
logs = state['log']
```

```python
print(logs.keys())
```

```python
train_loss, test_loss = logs['train_loss'], logs['test_loss']
train_error, test_error = logs['train_error'], logs['test_error']
train_acc, test_acc = [1-e for e in train_error], [1-e for e in test_error]
```

```python
axes = np.arange(0,len(train_acc))
plt.figure(figsize=(6,5))
plt.plot(axes, train_loss, 'k', label='train loss')
plt.plot(axes, test_loss, 'r--', label='val loss')
plt.title('Loss Trend for 60-Day Data')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(-0.2,3.5)
# plt.grid()
plt.legend()
plt.show()
```

```python
axes = np.arange(0,len(train_acc))
plt.figure(figsize=(6,5))
plt.plot(axes, train_acc, 'k', label='train acc.')
plt.plot(axes, test_acc, 'r--', label='val acc.')
plt.title('Predicting Accuracy for 60-Day Data')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
# plt.grid()
plt.legend()
plt.show()
```
