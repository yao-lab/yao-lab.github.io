```python
import numpy as np
import numpy.linalg as alg
import matplotlib.pyplot as plt
import pandas as pd
```

Compute the largest eigenvalue and its associated
eigenvector, with a comparison to the true ones.

```python

p = 100
n = 600
gamma = p/n

u = np.ones((p,1))  # true signal component
Ip = np.eye(p)
sigma_xs = np.arange(1,10,2)/10   # sigma^2
sigma_e = 1  # sigma^2
S = np.zeros((p,p))

for i,sigma_x in enumerate(sigma_xs):
    for j in range(n):
        alpha = np.random.normal(loc=0, scale=np.sqrt(sigma_x), size=1)
        epsilon = np.random.multivariate_normal(np.zeros(p), sigma_e * Ip, 1)
        x = alpha * u + epsilon
        S += 1/n * np.dot(x, x.T)   # cov matrix of (X+epsilon)

    eigenvalues, eigenvectors = alg.eig(S)
    value1 = eigenvalues[0]
    vector1 = eigenvectors[0]

    print('SNR=',sigma_x/sigma_e, 'sqrt(gamma)=',np.round(np.sqrt(gamma),2))
    if sigma_x/sigma_e > np.sqrt(gamma):
        print('SNR > sqrt(gamma)')
#         print('Theoretical principle eigenvalue=',(1+sigma_x)*(1+gamma/sigma_x))  # when sigma_e=1
    else:
        print('SNR <= sqrt(gamma)')
#         print('Theoretical principle eigenvalue=',(1+np.sqrt(gamma))**2)  # when sigma_e=1
    
#     print('The true signal component u:',u.T)
    print('Simulated principle eigenvalue:',value1)
    print('Second eigenvalue:', eigenvalues[1])
#     print('Corresponding eigenvector v:',vector1)
    print()
    
```
