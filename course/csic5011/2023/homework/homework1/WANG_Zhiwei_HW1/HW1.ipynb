# PCA experiments

```r
# Set up data matrix
digit <- fread("train.0", data.table = F)
X <- t(as.matrix(digit))
dim(X) # p = 256, n = 1194
p <- dim(X)[1]
n <- dim(X)[2]
head(X)
```

```r
# Compute the sample mean and center the data
mu <- rowMeans(X)
X_tilde <- X
```

```r
# top k SVD
X_svd <- svd(X_tilde)
X_svd$d[1:5] # top 5 singular value
k <- 50
X_hat <- X_svd$u[, 1:k] %*% diag(X_svd$d[1:k]) %*% t(X_svd$v)[1:k, ] # top k SVD
```

```r
# eigenvalue curve
Sig <- X_hat %*% t(X_hat) / n
Sig_eig <- eigen(Sig)
ggplot(data = NULL, aes(x = 1:k, y = Sig_eig$values[1:k] / sum(diag(Sig)))) +
    geom_point() +
    geom_line() +
    theme_bw() +
    theme(
        text = element_text(size = 20)
    )
```

# MDS of cities

```r
# Data matrix
cities <- read.csv("CityDistance.csv", header = TRUE)
D <- data.matrix(cities)
D
```

```r
# My own codes of MDS
MDS <- function(D, k = 2){
    n <- dim(D)[1]
    if(k > n){
        message("k is larger than n!")
    }
    H <- diag(n) - matrix(1, n, n)
    
    K_hat <- - H %*% D %*% H / 2
    K_hat_eigen <- eigen(K_hat)
    
    return(list(V = K_hat_eigen$vectors[, 1:k], lam = K_hat_eigen$values[1:k]))
}
```

```r
k <- 7
D_MDS <- MDS(D, k)
```

```r
# Plot the normalized eigenvalues
ggplot(data = NULL, aes(x = c(1:k), y = D_MDS$lam / sum(D_MDS$lam))) +
    geom_point() +
    geom_line() +
    theme_bw() +
    theme(
        text = element_text(size = 20)
    )
```
