## Problem 3: Finite rank perturbations of random symmetric matrices

### (a) Confirm Wigner's semi-circle law

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 3600
W = np.zeros((n,n))
W[np.triu_indices(n)] = np.random.randn(int(n*(n+1)/2)) /(2 * np.sqrt(n))
W += W.T + np.diag(W.diagonal())
W.shape
```

```python
lmd, V = np.linalg.eigh(W)
```

```python
n, bins, patches = plt.hist(lmd, density=1, bins=100)
y = 2 / np.pi * np.sqrt(1 - bins ** 2)
plt.plot(bins, y, '--', color ='black') 
plt.title("Eigenvalues Distribution")
plt.xlabel("eigenvalue")
plt.ylabel("frequency")

plt.show()
```
