```python
def load_meta_img(fn):
    reader =sitk.ImageFileReader()
    reader.SetImageIO("MetaImageIO")
    reader.SetFileName(fn)
    image = reader.Execute()
    img = list(np.int16(sitk.GetArrayViewFromImage(image)))
    return img

def mask_to_boundary_pts(mask):
    boundary_pts_list = []
    for i, slide in enumerate(mask):
        # check the whether the slide contains the patella mask 
        # find the slides which contain 2 elements (0 and 1)
        if len(np.unique(slide)) == 1:
            pass
        elif len(np.unique(slide)) == 2:
            boundary_pts = measure.find_contours(slide)[0]
            zcoord = np.ones(len(boundary_pts))*i # calculate the z-coordinates of the contour
            zcoord = zcoord.reshape(len(zcoord),1) # reshape z-coordinate list into (h,1)
            boundary_pts = np.hstack((boundary_pts, zcoord))
            boundary_pts_list.append(boundary_pts)
    
    surface_pts = np.vstack(boundary_pts_list)
    return surface_pts
```

```python
def tda_feat_extract(shape_data, repre_model_params:dict, plot_persistence=True):
    
    # Calculating single tda representation vector
    def tda_single_feat(repre, multi_pers, kth_pers):
        acX = gd.AlphaComplex(points=shape_data).create_simplex_tree()
        dgmX = acX.persistence()

        if repre == 'poly':
            CP = gr.ComplexPolynomial()
        elif repre == 'sil':
            CP = gr.Silhouette(resolution=100)
        elif repre == 'entropy':
            CP = gr.Entropy(mode='scalar')
        elif repre == 'landscape':
            CP = gr.Landscape() 
        elif repre == 'pi':
            CP = gr.PersistenceImage(bandwidth=1.0)

        if multi_pers == True:
            persistence_0th = acX.persistence_intervals_in_dimension(0)
            persistence_1st = acX.persistence_intervals_in_dimension(1)
            persistence_2nd = acX.persistence_intervals_in_dimension(2)
            persistence_0th[persistence_0th ==inf] = 0
            persistence_1st[persistence_1st ==inf] = 0
            persistence_2nd[persistence_2nd ==inf] = 0

            # Representation of 0th persistence
            CP.fit([persistence_0th])
            cp_0 = CP.transform([persistence_0th])
            cp_0 = cp_0.real.flatten()

            # Representation of 1st persistence
            CP.fit([persistence_1st])
            cp_1 = CP.transform([persistence_1st])
            cp_1 = cp_1.real.flatten()
            
            # Representation of 2nd persistence
            CP.fit([persistence_2nd])
            cp_2 = CP.transform([persistence_2nd])
            cp_2 = cp_2.real.flatten()

            single_tda = np.hstack((cp_0, cp_1, cp_2)).flatten()

        else:
            persistence = acX.persistence_intervals_in_dimension(kth_pers)
            persistence[persistence == inf] = 0
            CP.fit([persistence])
            cp = CP.transform([persistence])
            single_tda = cp.real.flatten()

        return single_tda, dgmX

    # Stack all tda representations horizontally to form a multi-tda vector
    if len(repre_model_params['representation']) > 1:
        b = []
        for repre in list(zip(*repre_model_params.values())):
            a, dgmX = tda_single_feat(*repre)
            b.append(a)
        X_tda = np.hstack(b)

    else:
        single_tda = tda_single_feat(*list(zip(*repre_model_params.values()))[0])
        X_tda, dgmX = single_tda
    
    if plot_persistence == True:
        gd.plot_persistence_diagram(dgmX[3:], max_intervals=100)
        #gd.plot_persistence_density(dgmX, dimension=1)
    return X_tda
```

```python
def tda_batch_extract(repre_model_params):
    DIR_tka = '../CT_scan_oa/Revised_CT_scan&clinical_data/CT_scan/TKA'
    DIR_uka = '../CT_scan_oa/Revised_CT_scan&clinical_data/CT_scan/UKA'

    paths_tka = [path for path in glob.glob(os.path.join(DIR_tka, '*/*/*.mha'))]
    paths_uka = [path for path in glob.glob(os.path.join(DIR_uka, '*/*/*.mha'))]

    df_tka = pd.DataFrame({'Group': np.zeros(len(paths_tka)),
                           'ID': list(range(len(paths_tka))),
                           'Path': paths_tka,
                           'TopoFeature': np.ones(len(paths_tka))})
    df_uka = pd.DataFrame({'Group': np.ones(len(paths_uka)),
                           'ID': list(range(len(paths_uka))),
                           'Path': paths_uka,
                           'TopoFeature': np.ones(len(paths_uka))})
    df = pd.concat([df_tka, df_uka], axis=0)
    df['TopoFeature'] = df['TopoFeature'].astype(object)

    @dask.delayed
    def run_extraction(path):
        mask = load_meta_img(path)
        pc = mask_to_boundary_pts(mask)
        x = tda_feat_extract(pc, repre_model_params)
        return x

    delayed_tda_list = []
    for path in df.Path:
        delayed_tda_list.append(
                                run_extraction(path)
                                    )

    with performance_report(filename="dask-report.html"):
        res = dask.delayed()(delayed_tda_list)
        pbar = ProgressBar() # register global progressbar
        pbar.register()
        tda_feat_list = res.compute()
    
    feature_names = ['TopoFeature'+str(i) for i in range(list(tda_feat_list.shape[1]))]
    df[feature_names] = tda_feat_list
    df.to_csv('TDA_features.csv')
    
    return df
```

```python
tda_params = {'representation': ['landscape'],
              'multi_pers': [True],
              'kth_pers':[0]}
df = tda_batch_extract(tda_params)
```

```python
X = np.vstack(np.array(df['TopoFeature'].values))
y = df['Group']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pca = PCA(n_components=25)
#smt = SMOTE(random_state=42)
#lr = LinearSVC()
lr = LogisticRegression()
#ens = EasyEnsembleClassifier(base_estimator=lr)
ens = SelfPacedEnsembleClassifier(base_estimator=lr)
#ens = RUSBoostClassifier(base_estimator=lr)
pipe = Pipeline(steps=[("pca", pca), ("clf", ens)])
pipe.fit(X_train, y_train)

y_tda_pred = pipe.predict(X_test)
print(metrics.classification_report(y_test, y_tda_pred))
print(metrics.roc_auc_score(y_test, y_tda_pred))
metrics.plot_confusion_matrix(pipe, X_test, y_test, normalize='true') 
```
