```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
genre_df = pd.read_csv("data/genre.csv")
Z= pd.read_csv("data/vfactor_proposed.csv")
```

```python
from sklearn import linear_model

fig, axes = plt.subplots(1,3, figsize=(30,15), sharex=True)
axes = axes.flatten()
for i in range(0,2):
    mlr = linear_model.LinearRegression()
    mlr.fit(genre_df, Z[:,i])
    
    ax = axes[i]
    ax.set_title(f"Factor {i+1}", fontdict={"fontsize":30})
    ax.bar(genre_df.columns, height=mlr.coef_)
    ax.tick_params(axis="both", which="major", rotation=90)
    for j in "top right left".split():
        ax.spines[j].set_visible(False)
        
fig.suptitle("Genre Loadings of Factors", fontsize=40)
plt.show()
```
