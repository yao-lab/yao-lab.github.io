# 1. Pre-processing

```python
import scTDA
```

```python
files = []
cells = []
libs = []
days = []
with open('data.txt', 'r') as f:
    for line in f:
        sp = line[:-1].split('\t')
        files.append(sp[0])
        cells.append(int(sp[1]))
        libs.append(sp[2])
        days.append(int(sp[3]))
```

```python
p = scTDA.Preprocess(files, days, libs, cells)
```

```python
p.show_statistics()
```

```python
p.subsample(8000)
```

```python
p.show_statistics()
```

```python
p.fit_sigmoid()
```

```python
p.select_genes(avg_counts=2.5, min_z=3)
```

```python
print p.genes[p.which_genes]
```

```python
p.save('Human_PFC_V4')
```
