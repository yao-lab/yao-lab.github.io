```python
from time import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
from sklearn import manifold
from scipy.io import loadmat    # to load matlab data


#Data proposessing
X = loadmat("./face.mat")   #Load data
Y = X['Y']
Y = np.reshape(Y,(10304, 33))
Y = Y.T
Y = np.float64(Y)  
```

```python
n_points = 33
n_neighbors = 5
n_components = 2
```

```python
#MDS Embedding of Face

mds = manifold.MDS(n_components, max_iter=2000, n_init=1)
Y1 = mds.fit_transform(Y)
plt.scatter(Y1[:, 0], Y1[:, 1], cmap=plt.cm.Spectral)
plt.title("MDS ")
plt.axis('tight')
```

```python
#Ordering face using MDS
Y1sort = np.argsort(Y1[:,0])
plt.figure(figsize=(20,20))
j=1
for i in Y1sort:
    t = Y[i].reshape(112, 92)
    plt.subplot(1,33,j,frameon=True)
    plt.imshow(t)
    j=j+1
    plt.axis('off')
plt.axis('off')
plt.figure()
```

```python
#ISOMAP
Y2 = manifold.Isomap(n_neighbors, n_components).fit_transform(Y)
t1 = time()
plt.scatter(Y2[:, 0], Y2[:, 1])
plt.title("Isomap") 
plt.axis('tight')
```

```python
#Ordering face using ISOMAP
Y2sort = np.argsort(Y2[:,0])
plt.figure(figsize=(20,20))
j=1
for i in Y2sort:
    t = Y[i].reshape(112, 92)
    plt.subplot(1,33,j,frameon=True)
    plt.imshow(t)
    j=j+1
    plt.axis('off')
plt.axis('off')
plt.figure()
```

```python
#LLE
Y3 = manifold.LocallyLinearEmbedding(n_neighbors, n_components).fit_transform(Y)
plt.scatter(Y3[:, 0], Y3[:, 1], cmap=plt.cm.Spectral)
plt.title("LLE")
plt.axis('tight')
plt.figure()
```

```python
#Ordering face using ISOMAP
Y3sort = np.argsort(Y3[:,0])
plt.figure(figsize=(20,20))
j=1
for i in Y3sort:
    t = Y[i].reshape(112, 92)
    plt.subplot(1,33,j,frameon=True)
    plt.imshow(t)
    j=j+1
    plt.axis('off')
plt.axis('off')
plt.figure()
```
