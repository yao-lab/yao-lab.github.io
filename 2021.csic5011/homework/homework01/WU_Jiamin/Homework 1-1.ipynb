```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
```

```python
X=np.array(np.loadtxt('train.3.txt',delimiter=","))
```

```python
X.shape
```

```python
X_mean=X.mean(axis=0)
X_mean.shape
```

```python
X_tuta=X-math.e*np.transpose(X_mean)
```

```python
U, S, V = np.linalg.svd(X_tuta)
```

```python
sigma_n=1/len(X_mean)*X_tuta*np.transpose(X_mean)
sigma_n.shape
```

```python
sigma_U, sigma_S, sigma_V = np.linalg.svd(sigma_n)
```

```python
k=len(S)
x = np.arange(0, k, 1)
y = sigma_S/S
plt.plot(x,y)
plt.xlabel("i")
plt.ylabel("Eigenvalue")
plt.title('Eigenvalue curve')
plt.show()
```
