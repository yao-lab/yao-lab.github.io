### Load the Data

```r
library('paran')
X<-read.csv('snp452-data.csv')
head(X, 6)
```

### A. Take the log prices

```r
Y<-log(X)
head(Y, 6)
```

### B. Calculate the log Price Jump

```r
dY<-apply(Y, 2, diff)
head(dY, 6)
```

### C. Construct Covariance Matrix

```r
n = dim(dY)[1]
dYt = t(dY)
S <- 1/n * (dYt%*%dY)
```

### D. Compute the Eigenvlues and Eigenvectors of S

```r
y <- eigen(S)

eigen_values = y$val
eigen_vectors = y$vec

de_eigen_values = sort(eigen_values, decreasing=TRUE)

# Show first 10 eigenvalues in decreasing order
de_eigen_values[1:10]
```

### E. Horn's Parallel Analysis
I use iterations = 500 not larger here because my computer is not so good. It breaks with large iteration number. 

```r
snp.paran<-paran(dY, iterations=500)
```

```r
snp.paran1<-paran(dY, iterations=50, centile=95)
```

There are some elements for snp.para and snp.para1:

- Retained: 
a scalar integer representing the number of components/factors retained
- AdjEv: 
a vector of the estimated eigenvalues adjusted for a finite sample size
- Ev: 
eigenvalues a vector of the eigenvalues of the observed data from either an unrotated principal component analysis or an unrotated common factor analysis
- RndEv: 
a vector of the estimated (mean or centile) eigenvalues from iterations number of N by P random data sets
- Bias:
a vector of the estimated bias of the unadjusted eigenvalues (i.e. the difference
between the adjusted and unadjusted eigenvalues)
- SimEvs: 
an iterations by P matrix with each row containing the eigenvalues from an
equivalent principal component or common factor analysis on an N by P data
set of uncorrelated random data
