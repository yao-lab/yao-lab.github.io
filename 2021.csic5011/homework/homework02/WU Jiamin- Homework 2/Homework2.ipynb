```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u=np.exp(1)
sigma=1
n=10
p=3
```

```python
def generate_data(sigma,u,n,p,lam_0):
    cov=sigma**2*np.identity(p)+lam_0*np.dot(u, u.T)
    x = np.random.multivariate_normal(np.zeros(p), cov, (n,n))
    S_n=0
    for i in range(0, n):
        S_n=S_n+np.dot(x[i], x[i].T)
    S_n=1/n*S_n
    return S_n
```

```python
y=[]
eigvec=[]
lam_true_ve=[]
for lam_0 in np.arange(0.5, 1.5, 0.1):
    lam_true=(sigma**2+lam_0)*(1+(p/n)*sigma**2/lam_0)
    S_n=generate_data(sigma,u,n,p,lam_0)
    lam, v= np.linalg.eig(S_n)
    y.append(lam[0])
    eigvec.append(v[0])
    lam_true_ve.append(lam_true)
print('eigenvector: ', eigvec)
x=np.arange(0.5, 1.5, 0.1)
plt.plot(x,y)
```

```python
plt.plot(lam_true_ve,y)
```
